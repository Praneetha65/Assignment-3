```python
# First we need to import necessary libraries
import pandas as pd  # Pandas are used for handling data
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
salary_data = pd.read_csv('/content/Salary_Data.csv')
print("Dataset loaded successfully.")

# Separate the data into input and output
X = salary_data[['YearsExperience']].values  # Input feature
y = salary_data['Salary'].values  # Output target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
print("Data split into training and testing sets.")

# Train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)
print("Model trained successfully.")

# Predict the salaries for the test data
y_pred = model.predict(X_test)
print("Predictions made on the test data.")

# Calculate the Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# Visualize the results and Training Data
plt.figure(figsize=(14, 7))

# Create the first subplot
plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, color='red', label='Actual (Training Data)')
plt.plot(X_train, model.predict(X_train), color='blue', label='Regression Line')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()

# Visualize Test Data
plt.subplot(1, 2, 2)
plt.scatter(X_test, y_test, color='green', label='Actual (Test Data)')
plt.plot(X_train, model.predict(X_train), color='blue', label='Regression Line')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()

plt.tight_layout()
plt.show()
```
